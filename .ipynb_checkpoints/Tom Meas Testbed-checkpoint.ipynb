```python
# simulating measurements, see "15 minutes to Qcodes"

class dmm_parameter(qc.Parameter):
    def __init__(self, name):
        super().__init__(name)
        self.ed = self.exponential_decay(5, 0.2)
        next(self.ed)

    def get_raw(self):
        """
        This method is automatically wrapped to
        provide a ``get`` method on the parameter instance.
        """
        val = self.ed.send(dac.ch1())
        next(self.ed)
        return val

    @staticmethod
    def exponential_decay(a: float, b: float):
        """
        Yields a*exp(-b*x) where x is put in
        """
        x = 0
        while True:
            x = yield
            yield a*np.exp(-b*x) + 0.02*a*np.random.randn()
            
#     def multiplier(A: float):
#         """
#         Multiplies a result by a constant A
#         """


dmm.v1 = dmm_parameter('dmm_v1')
```

```python
# Simulated Bscan where Vout (a simulated exp decay) is multiplied by set_B. Built using elements from "15 minutes to qcodes"

meas.write_period = 2

with meas.run() as datasaver:
    for set_B in np.linspace(0,10,10):
        Bsrc0.bigstep.set(set_B)
        #get_B = Bmeas0.bigmeas.get()
        for set_v in np.linspace(0, 20, 10):
            dac.ch1.set(set_v)
            get_v = dmm.v1.get()*set_B
            datasaver.add_result((Bsrc0.bigstep, set_B),
                                 (dac.ch1, set_v),
                                 (dmm.v1, get_v))
dataset = datasaver.dataset
```
